# Array Indexing and Slicing
1. Create a NumPy array of shape (6, 6) with values from 1 to 36. Extract the sub-array consisting of the 3rd to 5th rows and 2nd to 4th columns.
2. Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border. (HW)

```python
a = np.random.randint(1, 36, (6, 6))
print(a)
```

```python
print(a[2:5, 1:4])
```

```python
a = np.random.randint(1, 36, (6, 6))
b = np.random.randint(1, 36, (6, 6))

print(a[1], b[1])
```

```python
np.concatenate((a[1], b[1]))
```

```python
x = np.random.randint(1, 36, (5, 5))
print(x)
```

```python
top = a[0, :]
bottom = a[-1, :]
left = a[1: -1, 0]
right = a[1: -1, -1]

border = np.concatenate((top, right, bottom[::-1], left[::-1]))
print(border)
```

# Statistical Operations
1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array. ( HW)

```python
a = np.random.randint(1,4, (5, 5))
print(a)

print("Mean : ",np.mean(a))
print("Median : ", np.median(a))
print("Standard : ", np.std(a))
print("Variance : ", np.var(a))
```
