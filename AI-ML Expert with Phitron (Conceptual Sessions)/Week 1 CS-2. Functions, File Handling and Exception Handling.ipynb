# Map
map(function, iterable, ...)
Examples of Iterables
1. Lists: [1, 2, 3]
2. Tuples: (1, 2, 3)
3. Strings: "hello"
4. Dictionaries: {'a': 1, 'b': 2}
5. Sets: {1, 2, 3}
6. Files: open('file.txt')

```python
arr = [1, 2, 3, 4, 5, 6] # [10, 20, 30, 40, 50, 60]

def fun1(x):
    return x*10

arr2 = list(map(fun1, arr))
print(arr2)
```

```python
arr = [1, 2, 3, 4, 5, 6] # [10, 20, 30, 40, 50, 60]

arr2 = list(map(lambda x: x*10, arr))
print(arr2)
```

```python
scores = [85, 92, 78, 65, 45]

def grade(x):
    if x >= 80 : return f"Score is {x} and A+"
    if x >= 70 : return f"Score is {x} and A+"
    if x >= 60 : return f"Score is {x} and A-"
    else : return f"Score is {x} and F"


ans = list(map(grade, scores))
print(ans)
```

# Filter
filter(function, iterable)

```python
scores = [85, 92, 78, 65, 45]

ans = list(filter(lambda x : x>80, scores))
print(ans)
```

```python
scores = [85, 92, 78, 65, 45]

ans = list(map(lambda x : x>80, scores))
print(ans)
```

# Reduce 
reduce(function, iterable)

```python
from functools import reduce

scores = [85, 92, 78, 65, 45]

sum_scores = reduce(lambda x, y: x+y, scores)
print(sum_scores)
```

# Using Context Manager = With

```python
with open('sample.txt', 'r') as file:
    print(file.read())
print(file.closed)
```

```python
with open('simple.txt', 'r') as file:
    print(file.read(5))
```

```python
# Benefit => Load data by chunk

big_data = ["Hello AI\n" for i in range(100)]

with open('simple3.txt', 'w') as file:
    file.writelines(big_data)
```

```python
with open('simple3.txt', 'r') as file:
    print(file.read())
```

```python
with open('simple3.txt', 'r') as file:

    chunk_size = 10
    chunk_data = file.read(chunk_size)

    while len(chunk_data) > 0:
        print(chunk_data)
        chunk_data = file.read(chunk_size)
```
