```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
sale = pd.read_csv("sales.csv")

sale.head()
```

## 1. Using sales, draw a lineplot of date vs sales.

```python
sns.lineplot(data = sale, x = 'date', y = 'sales', errorbar=None)

plt.title('Data vs Sales')
plt.show()
```

```python
sns.lineplot(data = sale, x = 'sales', y = 'date', errorbar=None)

plt.title('Data vs Sales')
plt.show()
```

```python
student = pd.read_csv("students.csv")

student.head()
```

## 2. From students, plot hours_studied vs marks and color by gender.

```python
sns.scatterplot(data = student, x = 'hours_studied', y = 'marks', hue='gender')
```

```python
sns.relplot(kind = 'line', data = student, x = 'hours_studied', y = 'marks', hue='gender', errorbar=None)   # optional
```

## 3. Create a grid of scatterplots (hours_studied vs marks) faceted by subject.

```python
sns.relplot(data = student, x = 'hours_studied', y = 'marks', col='subject', kind='scatter')
```

## 4. Plot a histogram of marks with a KDE overlay.

```python
sns.histplot(data = student, x = 'marks', kde=True)
```

## 5. Show the KDE curve for hours_studied.

```python
sns.kdeplot(data = student, x = 'hours_studied')
```

## 6. Display the count of records per subject.

```python
student.head()
```

```python
sns.countplot(data = student, x = 'subject')
```

## 7. Draw a regression line for hours_studied and marks.

```python
sns.regplot(data = student, x = 'hours_studied', y = 'marks')
```

## 8. Create a pairplot for students' dataset. 

```python
students_performance = student[['hours_studied', 'marks', 'attendance_percentage', 'gender']]

students_performance
```

```python
sns.pairplot(data = students_performance, hue='gender')
```
