# File Pointer

```python
with open("sample.txt", "r") as file:
    print(file.tell())    # currently cursor koi ase
    print(file.read())
    print(file.tell())
    ## last index theke pora shuru korsi
    print(file.read())
```

```python
with open("sample.txt", "r") as file:
    print(file.tell())  
    print(file.read(5))
    print(file.tell())
    
    print(file.read())
```

```python
# seek

with open("sample.txt", "r") as file:
    print(file.tell())

    print(file.read(5))

    print(file.tell())

    file.seek(0)
    print(file.tell())
    print(file.read(5))
```

# Practice: 1
Get the text from sample.txt , and count the following

Number of lines

Number of words

Number of characters

and save it to the counter_of_string.txt file

```python
from functools import reduce

with open("sample.txt", "r") as file:
    string_list = file.readlines()

    # total line koi ta
    total_lines = len(string_list)

    # word ber korbo
    number_of_words = list(map(lambda x : len(x.split()), string_list))
    # total words
    total_number_of_words = reduce(lambda x, y: x+y, number_of_words)

    # cleaning process

    # 1. new line delete
    string_list = list(map(str.strip, string_list))

    # 2. space delete
    string_list = list(map(lambda x : x.replace(" ", ""), string_list))

    # koita character

    number_of_character = list(map(lambda x : len(x), string_list))

    # total character
    total_number_of_characters = reduce(lambda x, y : x+y, number_of_character)

    # new file
    with open("counter_of_string.txt", "w") as file:
        file.write(f"total line: {total_lines}\ntotal number of words: {total_number_of_words} \ntotal numbers of character: {total_number_of_characters}")
```
