# Handling Duplicates Values

```python
import pandas as pd
data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "David", "Bob"],
    "City": ["New York", "London", "Paris", "New York", "Tokyo", "London"],
    "Score": [85, 90, 78, 70, 95, 90]
}

df = pd.DataFrame(data)
df
```

```python
df.duplicated().sum()
```

```python
# deleting duplicates values

df.drop_duplicates()
```

```python
df
```

```python
# permanent

df.drop_duplicates(inplace=True)
```

```python
df
```

```python
# name er basis e duplicate deletion

df.drop_duplicates(subset=['Name'])
```

```python
df
```

```python
# city

df.drop_duplicates(subset=['Name', 'City'])
```

```python
# first duplicates value remove

df.drop_duplicates(subset=['Name'], keep='last')
```

```python
# last duplicate values remove

df.drop_duplicates(subset=['Name'], keep='first')
```

 # Apply Function in DF

```python
df
```

```python
# min max scaling

mn = df['Total Marks'].min()
mx = df['Total Marks'].max()

df['Scaled Marks'] = df['Total Marks'].apply(lambda x : (x-mn)/ (mx-mn))
```

```python
df
```

```python
# custom build function

def grading_system(marks):
    if marks > 260:
        return 'A+'
    elif marks >= 250:
        return 'A'
    else:
        return 'A-'

df['Grade'] = df['Total Marks'].apply(grading_system)
```

```python
df
```

```python

def marking_system(df):
    a = df['Data Structure Marks'] * 2
    b = df['Python Marks'] * 3
    c = df['Algorithm Marks'] * 4

    return a+b+c

df['Exceptional Marks'] = df.apply(marking_system, axis=1)
```

```python
df
```
