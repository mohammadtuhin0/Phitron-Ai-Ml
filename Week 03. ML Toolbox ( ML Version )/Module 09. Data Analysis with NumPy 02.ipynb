# concatenate

```python
com_row = np.concatenate((a, b), axis=0 ) # row wise

print(com_row)

con_col = np.concatenate((a, b), axis=1)   # column wise
print(con_col)
```

```python
# array split

a = np.random.randint(1, 10, size=(10, ))
print(a)

splitted_array = np.array_split(a,3)
print(splitted_array)

# split -> equal division
splitted_array = np.plit(a,3)
```

## Arithmetic operator and Mathematical functions

```python
# faster for vectorization

x = np.array([10, 20, 30, 40])
y = np.array([2, 3, 4, 5])

add = x+y
print(add)

sub = x-y
print(sub)

mul = x*y
print(mul)

div = x/y
print(div)

reminder = x%y
print(reminder)
```

```python
# mathematical function

# trigonometry

# sin, cos, tan takes values in radian
sin_val = np.sin(x)
print(sin_val)

cos_val = np.cos(x)
print(cos_val)

deg_con = np.rad2deg(x)
print(deg_con)

# deg2rad
rad_con = np.deg2rad(x)
print(rad_con)
```
