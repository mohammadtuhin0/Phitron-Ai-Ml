<h4>Max and Min Problem</h4>

```python
inp = input()
numbers = inp.split()

x = int(numbers[0])
y = int(numbers[1])
z = int(numbers[2])

min = x
max = y

if y < min:
    min = y

if z < min:
    min = z

if y > max:
    max = y
if z > max:
    max = z

print(min, max)
```

<h4>Even, Odd, Positive and Negative</h4>

```python
n = int(input())
inp = input()

numbers = inp.split()

positive = 0; negative = 0; even = 0; odd = 0

for i in range(n):
    x = int(numbers[i])
    if x > 0:
        positive += 1
    elif x < 0:
        negative += 1
    if x % 2 != 0:
        odd += 1
    else:
        even += 1


print("Even: ", even)
print("Odd: ", odd)
print("Positive: ", positive)
print("Negative: ", negative)
```

<h4>Digits</h4>

```python
t = int(input())

for i in range(t):
    number = int(input())
    if number == 0:
        print(0)
        continue
    while number > 0:
        print(number % 10, end=" ")
        number //= 10
    print()
```
