# List and LOOP

```python
nums = input().split()

even_sum = 0
even_count = 0

for num in nums:
    num = int(num)

    if num % 2 == 0:
        even_sum += num
        even_count += 1
```

```python
nums = list(map(int, input().split()))

even_sum = 0
even_count = 0

for num in nums:
    num = int(num)

    if num % 2 == 0:
        even_sum += num
        even_count += 1
```

# Dictionary

```python
numbers = input().split()

freq = {}

for num in numbers:
    num = int(num)
    
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)
```

# Generator

```python
def data_loader(batch_size, data):

    for i in range(0, len(data), batch_size):    # yield return chunk data, not whole data
        yield data[i : i+batch_size]

data = np.arange(20)

loader = data_loader(5, data)

print(next(loader))
print(next(loader))
print(next(loader))
print(next(loader))
```
