```python
class Dog:
    food = "Apple"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        print(f"{self.name} Dog speaking")

dog1 = Dog("dogName", 30)
dog2 = Dog("dogName2", 20)

print(dog1.food)
print(dog2.food)
```

```python
class Dog:
    food = "Apple"

    def __init__(self, name, age):
        self.name = name          # instance attribute
        self.age = age            # instance attribute

    def speak(self):              # instance method = instance attr, class attr
        print(f"{self.name} Dog speaking")

    @classmethod
    def work(cls):    #class method 
        print("Hello, dog ", self.food)

dog1 = Dog("dogName", 30)
dog2 = Dog("dogName2", 20)

dog1.speak()
dog2.work()
```
