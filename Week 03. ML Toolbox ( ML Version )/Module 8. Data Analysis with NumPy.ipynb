```python
arr = [1,2,3]

print(arr * 5)

arr = [x*5 for x in arr]
print(arr)
```

```python
import numpy as np

arr1 = np.array([1, 2, 3])
arr1 = arr1 * 5
print(arr1)
```

# Ndarray

```python
# 1d array

arr1 = np.array([1, 2, 3, 4, 5])
print(type(arr1))
print(arr1.ndim)

# 2d array
arr2 = np.array([[1, 2, 3],   # row1
                [4,5,6]])     # row2

print(arr2.ndim)

# 3d array
arr3 = np.array([
    # 1st floor
    [ [1, 2, 3],
    [4,5,6] ],
    
    # 2nd floor
    [ [1, 2, 3],
    [4,5,6] ],

    #3rd floor
    [ [1, 2, 3],
    [4,5,6] ],
])

print(arr3.ndim)
```

# Array Attributes

```python
# dimension
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

# shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

# data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

# size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

# ndarray data type

```python
arr = np.array([1, 2, 3])
print(arr.dtype)

arr = np.array([1.5, 2.4, 3.2])
print(arr.dtype)
```
