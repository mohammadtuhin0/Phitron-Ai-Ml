# Ndarray

```python
# 1d array

arr1 = np.array([1, 2, 3, 4, 5])
print(type(arr1))
print(arr1.ndim)

# 2d array
arr2 = np.array([[1, 2, 3],   # row1
                [4,5,6]])     # row2

print(arr2.ndim)

# 3d array
arr3 = np.array([
    # 1st floor
    [ [1, 2, 3],
    [4,5,6] ],
    
    # 2nd floor
    [ [1, 2, 3],
    [4,5,6] ],

    #3rd floor
    [ [1, 2, 3],
    [4,5,6] ],
])

print(arr3.ndim)
```

# Array Attributes

```python
# dimension
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

# shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

# data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

# size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

# array creation with range functions

```python
# np.arange(first, end, step) -> similar to range function

arr = np.arange(1, 10, 1)
mat = arr.reshape(3,3)

print(arr)
print(mat)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.linspace(start, end(included), koita)

arr = np.linspace(1, 10, 10)

print(arr)
print(mat)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
# np.logspace(start, end(included), koita)

arr = np.logspace(0, 4, 6, base=2)
print(arr)
```

# indexing and slicing

```python
# indexing

print(arr1)
print(arr2)

# arr1[2] = 100
# print(arr1)

print(arr2)
```

```python
# slicing
# 1d = arr[start:end:step]

arr1_mod = arr1[1:4].copy()
arr1[1] = 2

arr1_mod[2] = 200
print(arr1_mod)
print(arr1)
```

```python
# 2d slicing
# arr[row_start:row_end: step, column_start:column_end:step]

# print(arr2)
# getting a row

row_0 = arr2[0:1, ]
row_1 = arr2[1: , ]

print(row_0)
print(row_1)

# getting a column

col_0 = arr2[:: ,0:1]
col_1 = arr2[::, 1:2]

print(col_0)
print(col_1)

# getting a portion

portion = arr2[::, 0:2]
print(portion)
```
