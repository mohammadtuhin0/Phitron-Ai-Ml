# Converters

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv("cricket_matches.csv", converters={'team1':rename})
```

```python
pd.read_csv("cricket_matches.csv", na_values=["Abu Dhabi"])
```

# Export to Excel

```python
df1 = pd.read_json('https://jsonplaceholder.typicode.com/posts')
df2 = pd.read_json('https://jsonplaceholder.typicode.com/comments')
```

```python
with pd.ExcelWriter('post-comment.xlsx') as writer:
  df1.to_excel(writer, sheet_name="posts")
  df2.to_excel(writer, sheet_name="comments")
```
