Program 03: Write a simple program to calculate the area of a rectangle in Python

## The eval() Function

eval() function ববহােরর মােধম আমরা ইউজার থেক এ েপে ড ইনপুট িনেত পাির যমন এর আেগর পাগােমর দখেত পাি ইউজােরর থেক
কান সংখা ইনপুট নওয়ার সময় input() ফাংশনেক আমরা int() ফাংশেনর মেধ রাখেত হেয়িছেলা যােত কের আমরা ইউজার ইি জার টাইেপর ভালু
পাই এর কারণ হে আমরা যখন ইনপুট ফাংশন ববহার কির তখন এই ফাংশন ইউজােরর কাছ থেক িটং টাইেপর ইনপুট নয় তেব আমরা eval()
ফাংশন ববহার কের আমরা ইউজােরর কাছ থেক এ েপে ড ভালু িনেত পাির অথাৎ ইউজার যিদ সংখা ইনপুট দয় তাহেল আমরা সংখা পাব এবং
ইউজার যিদ কাের ার দয় তাহেল আমরা কাের ার টাইেপর ভালু পােবা

```python
number = eval(input("Enter a number: "))
string = input("Enter a string: ")

print(type(number))
print(type(string))
```
