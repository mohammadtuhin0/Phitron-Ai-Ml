```python
class Dog:
    # constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        print("Dog Speaking")

dog1 = Dog("DogName", 31)
dog2 = Dog("DogName2", 19)

print(dog1.name, dog2.name)
```

```python
class Dog:
    # constructor
    def __init__(self, name, age):
        self.name = name   # instance attribute
        self.age = age     # instance attribute

    def speak(self):       # instance attribute
        print(f"{self.name} Dog Speaking")

dog1 = Dog("DogName", 31)
dog2 = Dog("DogName2", 19)

dog1.speak()
dog2.speak()
```

```python
class Dog:
    food = "Apple"    # class attribute

    # constructor
    def __init__(self, name, age):
        self.name = name # instance attribute
        self.age = age

    def speak(self):
        print(f"{self.name} Dog speaking")

dog1 = Dog("DogName", 20)
dog2 = Dog("DogName2", 60)

print(dog1.food)
print(dog2.food)
```

```python
class Dog:
    food = "Apple"    # class attribute

    # constructor
    def __init__(self, name, age):
        self.name = name # instance attribute
        self.age = age

    def speak(self):
        print(f"{self.name} Dog speaking {self.food}")

    @classmethod
    def work(cls):          # class method = class attr
        print("Hello, dog", cls.food)

dog1 = Dog("DogName", 20)
dog2 = Dog("DogName2", 60)

dog1.speak()
dog1.work()
```

# Inheritance

```python
## Inherit (single inheritance)
## parent Class / Base class / Super class

class Car:
    def __init__(self, windows, door, enginetype) -> None:
        self.windows = windows
        self.door = door
        self.enginetype = enginetype

    def drive(self):
        print(f"The person will drive the {self.enginetype} card")

car1 = Car("", "", "")
dir(car1)
```

```python
# Derived / Child / subclass

class Tesla(Car):
    def __init__(self, windows, door, enginetype, is_selfDriving):
        super().__init__(windows, door, enginetype)
        self.is_selfDriving = is_selfDriving

tesla = Tesla(4, 2, "AB", True)
dir(tesla)
```

```python
class Farari(Car):
    def __init__(self, windows, door, enginetype, is_flying) -> None:
        super().__init__(windows, door, enginetype)
        self.is_flying = is_flying
```
