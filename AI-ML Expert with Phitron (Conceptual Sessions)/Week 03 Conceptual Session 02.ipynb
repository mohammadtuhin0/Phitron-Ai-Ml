# Mini Project

```python
data = {
    "Name" : ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age" : [20, 25, 23, 26, 29],
    "Marks" : [85, 67, 90, 45, 88]
}

df = pd.DataFrame(data)
df
```

```python
# add a grade columns

def assign_grade(marks):
    if marks >= 80:
        return "A"
    elif marks >= 60:
        return "B"
    else:
        return "C"

df["Grade"] = df["Marks"].apply(assign_grade)
```

```python
df
```

```python
# Filter Student

top_students = df[df["Marks"]>80]
top_students
```

```python
# sort student

sorted_df = df.sort_values(by="Marks", ascending=True)
sorted_df
```

```python
# average marks

avg_marks = df["Marks"].mean()
avg_marks
```

```python
# save as a csv file

df.to_csv("Student_output.csv", index=False)
```
