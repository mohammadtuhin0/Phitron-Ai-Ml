# Problem Solving-2 LeetCode

```python
import pandas as pd

def count_salary_categories(accounts: pd.DataFrame) -> pd.DataFrame:
    
    low_salary = (accounts['income']<20000).sum()
    average_salary = ((accounts['income']>=20000) & (accounts['income']<= 50000)).sum()
    high_salary = (accounts['income']>50000).sum()

    data = [('Low Salary', low_salary), ('Average Salary', average_salary),('High Salary', high_salary)]
    
    df = pd.DataFrame(data, columns=['category', 'accounts_count'])

    return df
```

# Filling Missing Data in Pandas-Problem Solve

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})

print(df)
```

```python
# Replace null values based on given conditions
df_fillna1 = df.fillna({'A': 0, 'B' : 99, 'C': 'Unknown'})

print(df_fillna1)
```

```python
# Forward fill column B
df_filled2 = df.copy()
df_filled2['B'] = df_filled2['B'].ffill()
print(df_filled2)
```

```python
# Fills missing values in numeric columns with the mean of the column and in string columns with 'Unknown'.

for column in df.columns:
    if df[column].dtype == 'object':
        df[column].fillna('Unknow', inplace=True)
    else:
        df[column].fillna(df[column].mean(), inplace=True)

print(df)
```
