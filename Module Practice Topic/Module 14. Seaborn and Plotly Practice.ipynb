```python
sale = pd.read_csv("sales.csv")

sale.head()
```

# 1. Using sales, draw a lineplot of date vs sales.

```python
student = pd.read_csv("students.csv")

student.head()
```

# 2. From students, plot hours_studied vs marks and color by gender.

```python
sns.lineplot(data = student, x = 'hours_studied', y = 'marks', hue='gender', errorbar=None)
```

```python
sns.relplot(kind = 'line', data = student, x = 'hours_studied', y = 'marks', hue='gender', errorbar=None)
```

# 3. Create a grid of scatterplots (hours_studied vs marks) faceted by subject.

```python
sns.scatterplot(data = student, x = 'hours_studied', y = 'marks', hue='subject')
```
