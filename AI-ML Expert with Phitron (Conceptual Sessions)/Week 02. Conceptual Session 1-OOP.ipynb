```python
class Dog:
    # constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        print("Dog Speaking")

dog1 = Dog("DogName", 31)
dog2 = Dog("DogName2", 19)

print(dog1.name, dog2.name)
```

```python
class Dog:
    # constructor
    def __init__(self, name, age):
        self.name = name   # instance attribute
        self.age = age     # instance attribute

    def speak(self):       # instance attribute
        print(f"{self.name} Dog Speaking")

dog1 = Dog("DogName", 31)
dog2 = Dog("DogName2", 19)

dog1.speak()
dog2.speak()
```

```python
class Dog:
    food = "Apple"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        print(f"{self.name} Dog speaking")

dog1 = Dog("dogName", 30)
dog2 = Dog("dogName2", 20)

print(dog1.food)
print(dog2.food)
```

```python
class Dog:
    food = "Apple"

    def __init__(self, name, age):
        self.name = name          # instance attribute
        self.age = age            # instance attribute

    def speak(self):              # instance method = instance attr, class attr
        print(f"{self.name} Dog speaking")

    @classmethod
    def work(cls):    #class method 
        print("Hello, dog ", self.food)

dog1 = Dog("dogName", 30)
dog2 = Dog("dogName2", 20)

dog1.speak()
dog2.work()
```
