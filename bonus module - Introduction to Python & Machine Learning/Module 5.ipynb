## Tuples

```python
# declaration

tup = (10, 20, 30, 40)
float_tup = (10.5, 10.2)
mixed_tup = (10, 10.5, "str", True)

lst = [10, 20, 30]

tup1 = tuple(lst)
print(type(tup1), tup1)
```

```python
# Access

tup = (10, 20, 30, 10.5, "str", False)

# index
print(tup[3])

# slicing
new_tup = tup[0:3]
print(new_tup)
```

```python
# mutable

lst = [10, 20, 30, 40]
lst.append(60)
lst[1] = 100
print(lst)

# immutable
tup = (10, 20, 30, 40)
tup[1] = 100
print(tup)
```

```python
# tuple method
tup = (10, 20, 30, 40, 10, 20, 30, 10)

print(tup.count(10))
print(tup.index(40))
```

## Set

```python
A = {1, 2, 3}
print(type(A), A)
```

```python
# empty structures
B = []
print(type(B))

C = ()
print(type(C))

S = set()
print(type(S))
```

```python
# accessing
S = {1, 2, 3, 4, 1, 2}
if (10 in S):
    print("10 in Present")
else:
    print("not present")

sum = 0
for element in S:
    sum += element
print(sum)
```

```python
# mutable
s = {5, 1, 2, 4, 3}
s.add(10)
s.add(0)

print(s)

s.pop()
s.pop()
print(s)

s.remove(5)
print(s)
```

## Set Mathematical Methods

```python
a = {1, 2, 3}
b = {1, 2, 5, 6, 4}

print(a.union(b))

print(a.intersection(b))

print(a.isdisjoint(b))

print(a.issubset(b))
```

## Dictionaries in Python

```python
# empty
dic = {}
print(type(dic))

dic = {
    "name" : "Tuhin",
    "age" : 20,
    "address" : "Gazipur",
    "numbers" : [10, 20, 30]
}

print(type(dic), dic)
```

```python
# access
# dic_name[key]

print(dic["numbers"])
print(dic.get("age"))

dic["name"] = "Phitron AI/ML"
dic = {
    "name" : "Phitron",
    "age" : 20,
    "adress" : "Dhaka",
    "numbers" : [10, 20, 30],
    "age" : 30
}

print(dic)
```

```python
# printing default values
print(dic)

print(dic.get("math_marks"))
print(dic.get("math_marks", 90))
```

```python
# adding a value
dic = {
    "name" : "Phitron",
    "age" : 20,
    "adress" : "Dhaka",
    "numbers" : [10, 20, 30],
    "age" : 30
}
print(dic)

dic["math_marks"] = 30
print(dic)
dic.update({
    "english_marks": 40
})
print(dic)
```

```python
# deleting
del dic["math_marks"]
print(dic)
```

```python
dic_2 = dic.copy()
print(dic_2)
```

## Iterating over Dictionary

```python
dic ={ "name" : "Phitron" , "age": 20 , "address" : "Dhaka" , "numbers":[10,20,30],"age":30}

keys = dic.keys()
print(keys)

dic["math_numbers"] = 30
print(keys)

values = dic.values()
print(values)

dic["math_numbers"] = 40
print(values)

items = dic.items()
print(items)
```

```python
# iteration
for key, value in dic.items():
    print(key, value)
```

## Dictionary Comprehension

```python
square = {x:x**2 for x in range(1, 11) if x%2==0}
print(square)
```

```python
Co_ordinates = [(10, 10.5), (20.5, 192), (101, 102)]
Locations = ["dhaka", "chattogram", "sylhet"]

exact_location = {co_or : loc for co_or, loc in zip(Co_ordinates, Locations)}
print(exact_location)
```

## Problem Solving with In Built Structures

1. Given a list of numbers, make a list with unique values

```python
lst = [10 , 20 , 10 , 30 , 30 , 50 , 30 , 10 , 20 , 10 , 10]

unique_values = set(lst)    #conversion
lst = list(unique_values)
print(lst)
```

2. Given a string , print the frequency of the words

```python
string = """data science machine learning data analysis machine 
learning statistics data models data training data validation features
features labels preprocessing data augmentation models data optimization 
gradient descent neural networks data tensors matrices visualization 
exploration pandas numpy matplotlib seaborn scikitlearn tensorflow pytorch 
deployment inference production monitoring reproducibility experiments results 
metrics accuracy precision recall f1 cross validation data machine
"""
words = string.split()
count = {}

for word in words:
    count[word] = count.get(word, 0) + 1

for k, v in count.items():
    print(f"count of {k} is {v}")
```
